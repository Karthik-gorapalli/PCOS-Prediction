```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Creating a contingency table (Example Data)
age_groups = ["20-30", "30-40","40-50"]
diseased = np.array([102, 68, 6])   # Number of diseased individuals
not_diseased = np.array([149,192,23])  # Number of healthy individuals

# Perform Chi-Square Test
data = np.array([diseased, not_diseased]).T
chi2, p, dof, expected = chi2_contingency(data)

# Calculate Disease Rate for Each Age Group
total_population = diseased + not_diseased
disease_rate = diseased / total_population  # Percentage of diseased individuals

# Find the age group with the highest disease rate
max_index = np.argmax(disease_rate)
most_affected_group = age_groups[max_index]

# Print Results
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:\n", expected)
print("\nDisease Rate by Age Group:")
for age, rate in zip(age_groups, disease_rate):
    print(f"{age}: {rate:.2%}")

print(f"\nMost affected age group: {most_affected_group} with a disease rate of {disease_rate[max_index]:.2%}")

# Interpretation
alpha = 0.05
if p < alpha:
    print("\nThere is a significant association between age group and disease status.")
else:
    print("\nNo significant association between age group and disease status.")
```
